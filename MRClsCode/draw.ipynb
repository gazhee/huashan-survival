```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# only MR
va_times = np.array([6,12,15,18,20,22,24,30,36])
cph_auc_huashan = np.array([0.8104575,0.8014768,0.8645647,0.8427518,0.8458798,0.8588033,0.8533727,0.8460897,0.8789048])
cph_mean_auc_huashan = cph_auc_huashan.mean()
cph_auc_test_TCGA = np.array([0.7942325, 0.8497818, 0.8392303, 0.8057803, 0.8583521, 0.8474561, 0.8396428, 0.8539825, 0.8430807 ])
cph_mean_auc_test_TCGA = cph_auc_test_TCGA.mean()
cph_auc = np.array([0.9590917, 0.9432931, 0.9479136, 0.9642604, 0.9434633, 0.9291413, 0.9414090, 0.9490596, 0.9539643 ])
cph_mean_auc = cph_auc.mean()
plt.plot(va_times, cph_auc,color= "#f9cc52",marker="o",label=f"Train TCGA Sequence (mean AUC = {cph_mean_auc:.3f})")
plt.plot(va_times, cph_auc_huashan,color="#e18283" ,marker="o",label=f"Huashan Sequence (mean AUC = {cph_mean_auc_huashan:.3f})")
plt.plot(va_times, cph_auc_test_TCGA,color="#0d898a" ,marker="o",label=f"Test TCGA Sequence (mean AUC = {cph_mean_auc_test_TCGA:.3f})")
plt.xlabel("months from enrollment", fontsize=14)
plt.ylabel("time-dependent AUC", fontsize=14)
plt.legend(loc="lower right")
plt.grid(True)
plt.xlim(3, 40)  # 设置横坐标范围
plt.ylim(0.5, 1)
plt.title('SE-DLFE-A Model', fontsize=18)
plt.savefig(r'G:\glioma\final_data\glioma_survive\only_mr\result\SE-DLFE-A-AUC.png',dpi = 500, bbox_inches="tight")
```
