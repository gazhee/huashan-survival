```python
import numpy as np
import pandas as pd
import os
```

```python
# 数据整合
TCGA_GBM_dir = '/root/autodl-tmp/huashan_MR/npy_file/TCGA_GBM_npy'
TCGA_LGG_dir = '/root/autodl-tmp/huashan_MR/npy_file/TCGA_LGG_npy'
pathology_TCGA_dir = '/root/autodl-tmp/huashan_MR/pathology/TCGA'
name = []
for patient in os.listdir(pathology_TCGA_dir):
    for file in os.listdir(TCGA_GBM_dir):
        if patient in file:
            name.append(patient)
    for file in os.listdir(TCGA_LGG_dir):
        if patient in file:
            name.append(patient)
```

```python
state = {
    'patient': name
}
df = pd.DataFrame(state)
df_clin = pd.read_csv('/root/autodl-tmp/MR_csv/glioma_survive/clin_all.csv')
df = pd.merge(df,df_clin,how = 'inner')
df.to_csv('/root/autodl-tmp/MR_csv/glioma_survive/TCGA_csv', index= False)
```

```python
huashan_dir = '/root/autodl-tmp/huashan_MR/npy_file/glioma_huashan_npy'
pathology_huashan_dir = '/root/autodl-tmp/huashan_MR/pathology/huashan'
name = []
for patient in os.listdir(pathology_huashan_dir):
    for file in os.listdir(huashan_dir ):
        if patient in file:
            name.append(patient)
state = {
    'patient': name
}
df = pd.DataFrame(state)
df_clin = pd.read_csv('/root/autodl-tmp/MR_csv/glioma_survive/clin_all.csv')
df = pd.merge(df,df_clin,how = 'inner')
df.to_csv('/root/autodl-tmp/MR_csv/glioma_survive/huashan_csv', index= False)
```
